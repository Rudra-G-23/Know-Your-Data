## 1.5 Multi-line Strings

```yaml
# Multi-line bio using '|' preserves line breaks
bio: |
  Hello, I am Rudra.
  I am learning machine learning.
  I want to work on industry-grade projects.

# Multi-line folded string using '>' turns newlines into spaces

  Hello, I am Rudra.
  I love learning machine learning.
  I also enjoy chess.
```

## 1.7 Null values & Booleans

```yaml

```


---

## 1.8 Explicit Quoting for tricky Values

```yaml
# YAML may treat "yes", "no", "on", "off" as booleans
# So we use quotes to force them as strings

status: "on"
confirmation: "yes"
```



---


## 1.9 Quick Summary of Rules

| ✅ What to do             | ❌ What to avoid                                |
| ------------------------ | ---------------------------------------------- |
| Use spaces, not tabs     | Don't mix tabs & spaces                        |
| Be consistent (2 spaces) | Avoid random indentation                       |
| Use `#` for comments     | Don’t write comments without `#`               |
| Quote tricky strings     | Don’t let `yes` turn into boolean accidentally |

## 2.1 Multi-line Strings (Literal `|` vs folder `>`)

```yaml
# Literal block scalar '|' preserves line breaks exactly
intro: |
  Hi, my name is Rudra.
  I love data science.
  This text will preserve line breaks.

# Folded block scalar '>' folds newlines into spaces
short_intro: >
  Hi, my name is Rudra.
  I love data science.
  This text will be folded into a single paragraph.
```

---

## 2.2 Anchors (&) and Aliases (*) - DRY principle
```yaml
# Using anchors & aliases to avoid repetition
defaults: &default_settings
  city: Bhubaneswar
  country: India
  active: true

profile:
  name: Rudra
  <<: *default_settings  # Merge the default settings here

friend:
  name: Abhi
  <<: *default_settings
  city: Cuttack  # override city just for friend
```


**How it works:**

- `&default_settings` defines an anchor named default_settings.

- `*default_settings` pulls it in wherever you write it.

- `<<: *default_settings` merges it like a copy-paste.
- This keeps your YAML DRY (Don't Repeat Yourself).

## 2.4 Explicit Data types

```yaml
# Forcing data types with tags
integer_value: !!int "42"
float_value: !!float "3.14159"
boolean_value: !!bool "true"
string_value: !!str 42  # stored as string, not int
```



- Most YAML parsers do auto detection, but forcing helps avoid mistakes.
---

## 2.5 YAML files with environment-specific overrides (like docker-compose)

```yaml
# docker-compose.yml
version: "3.8"
services:
  app:
    image: rudra/myapp:latest
    ports:
      - "5000:5000"
    environment:
      - ENV=production

# docker-compose.override.yml
services:
  app:
    environment:
      - ENV=development
```



---



## 🚀 Pro mini tips for intermediate YAML

- Always keep your **anchors close to where they’re used**, so files are easier to understand.
- When mixing nested lists & dicts, **indent consistently**, use 2 or 4 spaces — never tabs.
- Prefer multi-line strings (`|` or `>`) over giant inline strings for readability.
- Always test YAML merges with tools like `yq` or `docker-compose config`.

# ✍️ Pro Tips for writing YAML like a professional

- Always use spaces, never tabs.
- Most teams use 2 spaces. Consistency is more important than the number.
- Validate your YAML.
- Use yamllint or pyyaml to catch syntax errors.
- Keep lines short.
- No more than ~80-100 chars wide to stay readable.
- Use anchors (&) and aliases (*) to avoid repetition.
- Makes large files much cleaner.
- Add comments generously.
- YAML config files are read by humans; explain why a value is set.
- Be careful with yes, no, on, off.
- YAML treats them as booleans. Quote them if meant as strings: "no".
- Validate large configs against a schema.
- E.g. for Kubernetes: kubeval, kube-score. Format automatically.
- Use prettier --write '**/*.yml' or similar tools.
- Keep environment differences minimal.
- Use small override files instead of huge diffs for staging vs prod.
